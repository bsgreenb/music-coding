```python
# Test 1: Most basic matplotlib - should show 1 chart
import matplotlib.pyplot as plt
import numpy as np

plt.close('all')
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('Test 1: Basic plot')
plt.show()
print('Test 1 complete - should see 1 chart above')
```

```python
# Test 2: interact with 1 slider - should show 1 chart
from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt
import numpy as np

@interact(freq=FloatSlider(value=1, min=0.5, max=5, step=0.5, continuous_update=False))
def test_plot(freq):
    plt.close('all')
    x = np.linspace(0, 10, 100)
    plt.plot(x, np.sin(freq * x))
    plt.title(f'Test 2: freq={freq}')
    plt.show()

print('Test 2 complete - should see 1 chart above')
```

```python
# Test 3: interact with 3 sliders - should show 1 chart
from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt
import numpy as np

@interact(
    freq=FloatSlider(value=1, min=0.5, max=5, step=0.5, continuous_update=False),
    amp=FloatSlider(value=1, min=0.1, max=2, step=0.1, continuous_update=False),
    phase=FloatSlider(value=0, min=0, max=6.28, step=0.1, continuous_update=False),
)
def test_plot_3(freq, amp, phase):
    plt.close('all')
    x = np.linspace(0, 10, 100)
    plt.plot(x, amp * np.sin(freq * x + phase))
    plt.title(f'Test 3: freq={freq}, amp={amp}, phase={phase:.1f}')
    plt.show()

print('Test 3 complete - should see 1 chart above')
```
